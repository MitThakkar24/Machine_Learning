# Logistics Regression Implementation

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(dataset.data,columns = dataset.feature_names)
```

## Independent And Dependent Feature

```python
df['target'] = dataset.target
```

```python
df
```

```python
# Binary Classification 
df_copy = df[df['target']!=2]
```

## Divide in Dependent and independent feature

```python
X = df_copy.iloc[:,:-1]
y = df_copy.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20, random_state=42)
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train,y_train)
```

```python
# Prediciton 
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
## Confusion Matrix Accuracy Score, Classification report 
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
classifier.predict_proba(X_test)
```

```python
y_pred
```

## Hyperparameter Tuning

```python
## GridSearchCV
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
parameters = {'penalty':('l1','l2','elasticnet',None),
               'C':[1,10,20]
             }
```

```python
clf = GridSearchCV(classifier,param_grid=parameters,cv=5)
```

```python
# Splitting Train Data to Validation Data
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
classifier = LogisticRegression(C=1,penalty='l2')
```

```python
classifier.fit(X_train,y_train)
```

```python
# Prediction 
y_pred = classifier.predict(X_test)
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

## RandomizedSearchCv

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_clf = RandomizedSearchCV(LogisticRegression(),param_distributions=parameters,cv= 5)
```

```python
random_clf.fit(X_train,y_train)
```

```python
random_clf.best_params_
```
